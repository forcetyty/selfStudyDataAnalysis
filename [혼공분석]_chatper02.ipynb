#파이썬에서 JSON 데이터 다루기

```python
# 파이썬에서 JSON 데이터 다루기
# {"key":"value"} 형태의 데이터 포맷

d = {"name":"혼자 공부하는 데이터 분석"}
print(d['name'])
```

```python
# json.dumps() 함수
# 파이썬 객체를 JSON 문자열로 변환
# resure_ascii = False 사용 이유
# json.dumps() 함수는 아스키 문자 외의 다른 무자를 16진수로 출력하기 때문에 한글이 잘 표현안됨
# 따라서 한글 표기를 위해 사용함.

import json
d = {"name":"혼자 공부하는 데이터 분석"}
d_str = json.dumps(d,ensure_ascii=False)

# type 함수는 데이터 타입을 확인 할 수 있음
print(type(d_str))
```

```python
# json.loads() 함수
# JSON 문자열을 파이썬 객체로 변환
# 파이썬 객체를 바로 전송하지 않는 이유
# 웹 기반 API 통신 방식인 HTTP 가 텍스트 기반이기 때문에 객체를 HTTP 를 통해 전송할때 직렬화(serialization)이 필요
# 데이터를 받고 나서는 실행 가능하도록 역직렬화(deserialization)

import json
d = {"name":"혼자 공부하는 데이터 분석"}
d_str = json.dumps(d,ensure_ascii=False)

d2 = json.loads(d_str)
print(d2['name'])

# type 함수는 데이터 타입을 확인 할 수 있음
print(type(d2))
```

```python
import json
d3 = json.loads('{"name":"혼자 공부하는 데이터 분석", "author":"박해선", "year":2022}')

print(d3['name'])
print(d3['author'])
print(d3['year'])
```

#파이썬에서 XML 데이터 다루기

```python
# xml.etree.ElementTree 모듈의 fromstring() 함수를 이용하여 문자영을 파이썬 객체로 변환

import xml.etree.ElementTree as et

x_str = """
<book>
  <name>혼자 공부하는 데이터 분석</name>
  <author>박해선</author>
  <year>2022</year>
</book>
"""

# fromstring() 함수가 반환하는 객체는 단순한 파이썬 객체가 아니라
# ElementTree 모듈 아래에 정의된 Element 클래스의 객체
book = et.fromstring(x_str)

print(type(book))
print(book.tag)
```

```python
# findtext() 메소드
# 자식 엘리먼트 확인 하기

import xml.etree.ElementTree as et

x_str = """
<book>
  <name>혼자 공부하는 데이터 분석</name>
  <author>박해선</author>
  <year>2022</year>
</book>
"""

# fromstring() 함수가 반환하는 객체는 단순한 파이썬 객체가 아니라
# ElementTree 모듈 아래에 정의된 Element 클래스의 객체
book = et.fromstring(x_str)

book_childs = list(book)  # 리스트로 변환할 때는 list() 함수를 사용

print(book_childs)

# book_childs 리스트 각 항목을 name, author, year 변수에 할당하고 text 속성으로 엘리먼트에 있는 텍스트 출력
# book_childs 리스트에 속성값에 접근

# 단점, XML은 자식 엘리먼트 순서가 항상 일정하다는 것을 보장 하지 않음
name, author, year = book_childs
print(name.text)
print(author.text)
print(year.text)
```

```python
# findtext() 메소드
# 자식 엘리먼트 확인 하기

import xml.etree.ElementTree as et

x_str = """
<book>
  <name>혼자 공부하는 데이터 분석</name>
  <author>박해선</author>
  <year>2022</year>
</book>
"""

# fromstring() 함수가 반환하는 객체는 단순한 파이썬 객체가 아니라
# ElementTree 모듈 아래에 정의된 Element 클래스의 객체
book = et.fromstring(x_str)

book_childs = list(book)  # 리스트로 변환할 때는 list() 함수를 사용

print(book_childs)

# book_childs 리스트 각 항목을 name, author, year 변수에 할당하고 text 속성으로 엘리먼트에 있는 텍스트 출력
# book_childs 리스트에 속성값에 접근

# (1)
# 단점, XML은 자식 엘리먼트 순서가 항상 일정하다는 것을 보장 하지 않음
# name, author, year = book_childs
# print(name.text)
# print(author.text)
# print(year.text)

# findtext() 메소드를 사용하면 해당 자식 엘리먼트를 탐색하여 자동으로 텍스트르 반환할 수 있음
# (1)의 방법 보다 더 정확하고 안전하게 원하는 값을 가져 올 수 있음
name = book.findtext('name')
author = book.findtext('author')
year = book.findtext('year')

print(name)
print(author)
print(year)
```

```python
# findtext() 메소드
# 자식 엘리먼트 확인 하기

import xml.etree.ElementTree as et

x2_str = """
<books>
  <book>
    <name>혼자 공부하는 데이터 분석</name>
    <author>박해선</author>
    <year>2022</year>
  </book>
  <book>
    <name>혼자 공부하는 머시러닝+딥러닝</name>
    <author>박해선</author>
    <year>2020</year>
  </book>
</books>
"""

books = et.fromstring(x2_str)
print(books.tag)
```

```python
# findall() 메소드와 for 문

import xml.etree.ElementTree as et

x2_str = """
<books>
  <book>
    <name>혼자 공부하는 데이터 분석</name>
    <author>박해선</author>
    <year>2022</year>
  </book>
  <book>
    <name>혼자 공부하는 머시러닝+딥러닝</name>
    <author>박해선</author>
    <year>2020</year>
  </book>
</books>
"""

# books 객체에 있는 tag 속성에 접근하여 출력
# for문을 통해 순회하며 books 객체를 출력함
for book in books.findall('book'):
  name = book.findtext('name')
  author = book.findtext('author')
  year = book.findtext('year')

print(name)
print(author)
print(year)

# xml을 판다스로 출력
# pd.read_xml(x2_str)
```

```python
# 판다스 DataFrame 클래스에 리스트 넘기기
# requests 패키지
# 파이썬으로 API 호출하기

# 도서관 정보나루 API 키
# ULR은 HTTP GET 방식으로 파라미터 값을 전달
import requests
import pandas as pd

url = "http://data4library.kr/api/loanItemSrch?format=json&startDt=2021-04-01&endDt=2021-04-30&authKey=apikey"

r = requests.get(url)

#json 메소드는 웹서버로 받은 JSON 문자열을 파이썬 객체로 변환하여 반환됨
data = r.json()

# doc 키에 매핑된 딕셔너리를 추출한 뒤 빈 리스트에 추가
books = []
for d in data['response']['docs']:
  books.append(d['doc'])

# 파이썬의 리스트 내포로 만드는 방법
#books = [d['doc'] for d in data['response']['docs']]

books_df = pd.DataFrame(books)
books_df

# books_df 의 내용을 JSON으로 변환하여 저장
# to_json 메소드는 유니코드를 16진수로 저장하기 때문에 한글이 제대로 안타날수 있음
# 따라서, force_ascii=False 로 설정
books_df.to_json('20s_best_book.json', force_ascii=False)
```

# 실무 Tip
API 방식을 이용할 때 데이터 형식이 예기치 못하게 변경 될 가능성이 있다.</br>
따라서, 데이터 수집 프로그램을 작성하여 정지적으로 데이터를 가져오면서 API 호출에 오류가 나타나지</br>
않는지 체크가 필요하다.

## key Point
API : 프로그램 간 데이터를 전달하기 위해 정한 규칙 </br>
HTTP : 웹에서 데이터를 주고받기 위한 프로토콜 </br>
JSON : Javascript로 표기되는 데이터 포맷 </br>
XML: 확장형 마크업 언어 </br>

## 기억!!!
json.dumps() : 파이썬 객체를 JSON 문자열로 변환
json.loads() : JSON 문자열을 파이썬 객체로 변환
pandas.read_json() : JSON 문자열을 판다스 시리즈나 데이터프레임으로 변환
xml.etree.ElementTree.fromstring() : XML 문자열을 분석하여 xml.etree.ElementTree.Element </br> 클래스 객체를 반환 </br>
xml.etree.ElementTree.Element.findtext() : 지정한 태그 이름과 맞는 첫 번째 자식 엘리먼트의 텍스트를 반환 </br>
xml.etree.ElementTree.Element.findall() : 지정한 태그 이름과 맞는 모든 자식 엘리먼트 반환
</br>
requests.get() : HTTP GET 방식으로 URL을 호출하고 requests.Response 객체 반환</br>
requests.Response.json() : 응답받은 JSON 문자열을 파이썬 객체로 변환 </br>

```python
# HTML에서 데이터 추출하기 : 뷰티플수프
# HTML 에서 데이터 추출하는 툴
# (1) 뷰티블수프 Beautiful Soup
# (2) 스크래피 Scrapy

# HTML 가져오기 : requests.get() 함수
import requests
from bs4 import BeautifulSoup

# 전체 도서의 쪽수 구하기
# 판다스 데이터프레임은 한 행씩 순차적으로 처리하느데 최적화 되어 있지 않음.
# 데이터프레임을 for문으로 반복하는 것은 비효율적
# 대신, 데이터프레임은 각 행 또는 각 열에 원하는 함수를 자동으로 적용해 주는 여러가지 방법을 제공

def get_page_cnt(isbn):
  # Yes24 도서 검색 페이지 URL
  url = 'http://www.yes24.com/Product/Search?domain=BOOK&query={}'

  # URL에 ISBN을 넣어 HTML 획득
  r = requests.get(url.format(isbn))
  soup = BeautifulSoup(r.text, 'html.parser') #html 파싱

  #print(soup) #디버깅용

  # 검색 결과에서 해당 도서를 선택
  prd_info = soup.find('a', attrs={'class':'gd_name'})
  if prd_info == None:
    return ''

  #print(prd_info) #디버깅용

  # 도서 상세 페이지를 가져옴
  url = 'http://www.yes24.com'+prd_info['href']
  r = requests.get(url)
  soup = BeautifulSoup(r.text, 'html.parser')

  # 품목정보 <div>를 선택
  prd_detail = soup.find('div', attrs={'id':'infoset_specific'})

  # 테이블에 있는 <tr> 태그를 가져옴
  prd_tr_list = prd_detail.find_all('tr')

  # 쪽수가 들어 있는 <th>를 찾아 <td>에 담긴 값을 반환
  for tr in prd_tr_list:
    if tr.find('th').get_text() == '쪽수, 무게, 크기':
      return tr.find('td').get_text().split()[0]
  return ''

get_page_cnt(9791190090018)

```
